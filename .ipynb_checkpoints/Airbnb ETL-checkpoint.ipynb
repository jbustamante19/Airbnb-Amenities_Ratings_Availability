```python
#importing libraries
import pandas as pd
import sys 
import urllib
from splinter import Browser
from bs4 import BeautifulSoup
import time
from selenium.webdriver import ChromeOptions
```

```python
#options = {'chrome_options':['--disable-blink-features=AutomationControlled','--disable-setuid-sandbox']}

chrome_options = ChromeOptions()
#chrome_options.add_argument('--disable-blink-features=AutomationControlled')
#chrome_options.add_argument('start-maximized')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
browser=Browser('chrome',options=chrome_options,headless=False)
```

```python
#Launching automated browser & navigating to url of interest
#browser=Browser('chrome',headless=False)
url = 'https://www.airbnb.com/s/Real-del-Monte--Hgo.--Mexico/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&price_filter_input_type=0&price_filter_num_nights=1&channel=EXPLORE&date_picker_type=calendar&source=structured_search_input_header&search_type=autocomplete_click&query=Real%20del%20Monte%2C%20Hgo.%2C%20Mexico&place_id=ChIJdwSsTtgH0YUR3Y8E-Ohc8o4'
browser.visit(url)
time.sleep(5)
html = browser.html
#getting html script of the matches data and parsing it with beautiful soup
soup = BeautifulSoup(html,'html.parser')
#storing html with match information and link to more match details; found by searching for the class
homes=soup.find(class_= 'gsgwcjk').find_all(class_='cy5jw6o')
numberOfHomes = soup.h1.find_all('span')[0].text.split('results')[1].split(' ')[0]
print(numberOfHomes)


results = []
page=1
homeNumber = 1
while int(numberOfHomes) >= homeNumber:
    for home in homes:
        if homeNumber<=page*18 and int(numberOfHomes)>=homeNumber:
            print('Home number: ',homeNumber)
            print(home.a['href'])
            results.append(home.a['href'])
            print('----')
            homeNumber+=1
        elif int(numberOfHomes)>=homeNumber: 
            page+=1
            browser.visit('https://airbnb.com'+soup.find(class_="l1ovpqvx c1ytbx3a dir dir-ltr")['href'])
            time.sleep(4)
            html=browser.html
            soup = BeautifulSoup(html,'html.parser')
            homes=soup.find(class_= 'gsgwcjk').find_all(class_='cy5jw6o')
```

```python
nameOfLists=['name','placeType','city','state','country','rating','reviewsNumber','hostedBy','guestNum','bedroomNum','bedNum','bathNum','cleanliness','accuracy','communication','location','checkIn','Value','full_location','hostJoined','pricePerNight']
listOfLists=[]

for l in nameOfLists:
    l = list()
    listOfLists.append(l)
listOfLists
```

```python
place=0

stayDate=[]
stayAvail=[]
placeName= []
dateDetailed=[]

for result in results:
    place+=1
    print('Airbnb number: ',place)
    #print(home.a['href'])
    #print(type(home.a['href']))
    
    browser.visit('https://airbnb.com'+result)
    time.sleep(3)
    
    try:
        browser.find_by_xpath('/html/body/div[10]/section/div/div/div[2]/div/div[1]/button').click()
    except:
        print("No translation Message.")
    
    time.sleep(2)
    try:
        browser.find_by_xpath('//*[@id="site-content"]/div/div[1]/div[11]/div/div/div/div[2]/div[1]/div[2]/div/div[2]/div/button').click()
        #browser.find_by_xpath('//*[@id="site-content"]/div/div[1]/div[3]/div/div[1]/div/div[7]/div/div[2]/div[1]/div[2]/div/div[2]/div[2]/button').click()
    except:
        print("not cleared dates #1")
    try:
        browser.find_by_text('Clear dates').click()
        #browser.find_by_xpath('//*[@id="site-content"]/div/div[1]/div[11]/div/div/div/div[2]/div[1]/div[2]/div/div[2]/div/button').click()
        #browser.find_by_xpath('//*[@id="site-content"]/div/div[1]/div[3]/div/div[1]/div/div[7]/div/div[2]/div[1]/div[2]/div/div[2]/div[2]/button').click()
        print('Cleared Dates using text')                       
    except:
        print("error by text")
    time.sleep(2)
    html = browser.html
    detailsSoup = BeautifulSoup(html,'html.parser')
    
    try:
        for i in detailsSoup.find(class_='_2hyui6e').find_all("td",class_=True):
            month = i['aria-label'].split('.')[0].split(',')[2].split(' ')[1]
            #print(i['aria-label'].split('.')[0].split(',')[2].split(' ')[1])
            day = i['aria-label'].split('.')[0].split(',')[0]
            #print(i['aria-label'].split('.')[0].split(',')[0])
            year = i['aria-label'].split('.')[0].split(',')[2].split(' ')[2]
            #print(i['aria-label'].split('.')[0].split(',')[2].split(' ')[2])
            dateStay = month+' '+day+', '+year
            print(dateStay)
            datedetailed = i['aria-label'].split('.')[0]
            #print(date['aria-label'].split('.')[0])
            availibility = i['aria-label'].split('.')[1]
            print(availibility)
            print('------')
            stayDate.append(dateStay)
            stayAvail.append(availibility)
            placeName.append(detailsSoup.find_all(class_='hpipapi')[0].text)
            dateDetailed.append(datedetailed)
    except:
        print('error caught')
    
    try:
        name = detailsSoup.find_all(class_='hpipapi')[0].text
        listOfLists[0].append(name)
        print(name)
    except:
        print('Something wrong call, Anne Phong!-name')
        listOfLists[0].append(None)
        
    try:
        placeType = detailsSoup.find_all(class_='hpipapi')[1].text.split(" hosted")[0]
        listOfLists[1].append(placeType)
        print(placeType)
    except:
        print('Something wrong call, Anne Phong!-placetype')
        listOfLists[1].append(None)
        
        
    try:
        city = detailsSoup.find(class_="_152qbzi").text.split(', ')[0]#detailsSoup.find(class_="_leqb4t").text.split(', ')[0]
        print(city)
        state = detailsSoup.find(class_="_152qbzi").text.split(', ')[1]
        print(state)
        country = detailsSoup.find(class_="_152qbzi").text.split(', ')[2]
        print(country)
        listOfLists[2].append(city)
        listOfLists[3].append(state)
        listOfLists[4].append(country)
        
    except:
        print('Something wrong call, Anne Phong!-city,state,country')
        listOfLists[2].append('NA')
        listOfLists[3].append('NA')
        listOfLists[4].append('NA')
    try:
        entireLocation = detailsSoup.find(class_='_leqb4t').text
    except:
        entireLocation = 'NA'
        if entireLocation == 'NA':
            try:
                entireLocation = detailsSoup.find(class_='_152qbzi').text
            except:
                entireLocation = None
                print('Something wrong call, Anne Phong!-entirelocation')
    print(entireLocation)    
    listOfLists[18].append(entireLocation)
        
        
    
            
        
    try:
        rating = detailsSoup.find(class_="t1f90fvr dir dir-ltr").div.text.split(' · ')[0]
        listOfLists[5].append(rating)
        print(rating)
    except:
        print('Something wrong call, Anne Phong!-rating')
        listOfLists[5].append(None)
        
    try:
        reviewsNumber = detailsSoup.find(class_="t1f90fvr dir dir-ltr").div.text.split(' · ')[1].split(' ')[0]  
        listOfLists[6].append(reviewsNumber)
        print(reviewsNumber)
    except:
        print('Something wrong call, Anne Phong!-reviewNumber')
        listOfLists[6].append(None)
        
    try:
        hostedBy =detailsSoup.find_all(class_="hnwb2pb dir dir-ltr")[4].text.split('by ')[1]
        listOfLists[7].append(hostedBy)
        print(hostedBy)
    except:
        print('Something wrong call, Anne Phong!-hostedBy')
        listOfLists[7].append(None)
        
    try:
        guestsNum =detailsSoup.find('li',class_="l7n4lsf dir dir-ltr").text.split(' ')[0] #detailsSoup.find_all(class_="lgx66tx dir dir-ltr")[0].text.split(' ')[0]
        listOfLists[8].append(guestsNum)
        print(guestsNum)
    except:
        print('Something wrong call, Anne Phong!-guestNum')
        listOfLists[8].append(None)
        
    try:
        bedroomNum = detailsSoup.find_all(class_="lgx66tx dir dir-ltr")[0].text.split(' ')[5]
        listOfLists[9].append(bedroomNum)
        print(bedroomNum)
    except:
        print('Something wrong call, Anne Phong!-bedroomNum')
        listOfLists[9].append(None)
        
    try:
        bedNum = detailsSoup.find_all(class_="lgx66tx dir dir-ltr")[0].text.split(' ')[5]
        listOfLists[10].append(bedNum)
        print(bedNum)
    except:
        print('Something wrong call, Anne Phong!-bedNum')
        listOfLists[10].append(None)
        
    try:
        bathNum = detailsSoup.find_all(class_="lgx66tx dir dir-ltr")[0].text.split(' ')[15]
        listOfLists[11].append(bathNum)
        print(bathNum)
    except:
        print('Something wrong call, Anne Phong!-bathNum')
        listOfLists[11].append(None)
        
    try:
        cleanliness = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[0].text.split('Cleanliness')[1]
        listOfLists[12].append(cleanliness)
        print(cleanliness)
    except:
        print('Something wrong call, Anne Phong!-cleanliness')
        listOfLists[12].append(None)
        
    try:
        accuracy = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[1].text.split('Accuracy')[1]
        listOfLists[13].append(accuracy)
        print(accuracy)
    except:
        print('Something wrong call, Anne Phong!-accuracy')
        listOfLists[13].append(None)
        
    try:
        communication = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[2].text.split('Communication')[1]
        listOfLists[14].append(communication)
        print(communication)
    except:
        print('Something wrong call, Anne Phong!-communication')
        listOfLists[14].append(None)
        
    try:
        location = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[3].text.split('Location')[1]
        listOfLists[15].append(location)
        print(location)
    except:
        print('Something wrong call, Anne Phong!-location')
        listOfLists[15].append(None)
        
    try:
        checkIn = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[4].text.split('Check-in')[1]
        listOfLists[16].append(checkIn)
        print(checkIn)
    except:
        print('Something wrong call, Anne Phong!-checkIn')
        listOfLists[16].append(None)
        
    try:
        value = list(detailsSoup.find(class_="ciubx2o dir dir-ltr"))[5].text.split('Value')[1]
        listOfLists[17].append(value)
        print(value)   
    except:
        print('Something wrong call, Anne Phong!-value')
        listOfLists[17].append(None)
    try:
        hostJoined = detailsSoup.find(class_='s9fngse dir dir-ltr').text.split(' ')[3]
        listOfLists[19].append(hostJoined)
        print(hostJoined)
    except:
        print('Something wrong call, Anne Phong!-value')
        listOfLists[19].append(None)
    try: 
        pricePerNight = detailsSoup.find(class_='_tyxjp1').text.split('$')[1].split("\xa0")[0]
        listOfLists[20].append(pricePerNight)
        print(pricePerNight)
    except:
        print('Something wrong call, Anne Phong!-priceperNight')
        listOfLists[20].append(None)
        
```
